```python
import os
import torchaudio
import random
from pathlib import Path
import matplotlib.pyplot as plt

CLEAN_PATH = Path(os.getenv('CLEAN_DATA_PATH', '../data/raw'))
NOISY_PATH = Path(os.getenv('PROCESSED_DATA_PATH', '../data/processed'))

files = list(CLEAN_PATH.glob("*.wav"))

for _ in range(3):
    file = random.choice(files)
    waveform, sr = torchaudio.load(file)

    plt.figure(figsize=(14, 3))
    plt.title(f"Waveform: {file.name}")
    plt.plot(waveform.t().numpy())
    plt.xlabel("Samples")
    plt.ylabel("Amplitude")
    plt.show()

    print(f"Duration: {waveform.size(1)/sr:.2f} sec | Sample Rate: {sr}")
```
